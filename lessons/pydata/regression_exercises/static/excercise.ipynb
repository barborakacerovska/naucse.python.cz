```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("population.csv", index_col=0)
df.sample(6)
```

```python
%matplotlib inline

df.plot.scatter(
    x="Rozloha",
    y="Počet obyvatel"
);
df.plot.scatter(
    x="Rozloha",
    y="Počet obyvatel",
    xlim=[0,2000]
);
```

```python
X = df[["Rozloha"]]
y = df["Počet obyvatel"]
```

```python
X_train, y_train = X[:70], y[:70]
X_test, y_test = X[70:], y[70:]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
cerna_krabicka = LinearRegression()
cerna_krabicka.fit(X_train, y_train)

pred = cerna_krabicka.predict(X_test)
pred_train = cerna_krabicka.predict(X_train)

```

```python
fig, ax = plt.subplots(1, 2, figsize=(15,5))

ax[0].plot(X_test, pred, color="red")
ax[0].scatter(X, y, color="green");

ax[1].plot(X_test, pred, color="red")
ax[1].scatter(X, y, color="green")
ax[1].set_xlim([0,2000]);
```
